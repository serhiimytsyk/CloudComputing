```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

Import data, calculate percentiles, and plot them

```python
id = "iot_producer_one_1"
with open("../iot_producer/target/" + id + ".json", "r") as file:
    data = list(json.load(file).values())
    data.pop(0)
counts = np.arange(1, len(data) + 1)

p90 = np.percentile(data, 90)
p95 = np.percentile(data, 95)

plt.plot(counts, data, label='Latencies')
plt.axhline(y = p90, color='orange', label='P90')
plt.axhline(y = p95, color='red', label='P95')
plt.title('Latency Analysis with P90 and P95 Percentiles with One producer running')
plt.xlabel('Sample Count')
plt.ylabel('Latency')
plt.legend()
plt.show()
```

Compute CDF and plot it

```python
sorted_data = np.sort(data)
probabilities = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
plt.plot(sorted_data, probabilities)
plt.title('Cumulative Distribution Function (CDF) of Latencies with One producer running')
plt.xlabel('Latency')
plt.ylabel('Probability')
plt.show()
```
