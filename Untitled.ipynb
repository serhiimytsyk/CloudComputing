```python
from statsmodels.tsa.stattools import adfuller

# Perform ADF test
result = adfuller(mid_prices)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")

if result[1] < 0.05:
    print("The data is stationary.")
else:
    print("The data is not stationary. Differencing is required.")
```

```python
# Differencing to make the data stationary
mid_prices_diff = mid_prices.diff().dropna()

# Recheck stationarity after differencing
result = adfuller(mid_prices_diff)
print(f"p-value after differencing: {result[1]}")
```

```python
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

def get_arima_coefficients(dataset, p, d, q):
    model = ARIMA(dataset, order=(p, d, q))
    fitted_model = model.fit()
    
    arparams = fitted_model.arparams
    maparams = fitted_model.maparams
    intercept = fitted_model.params[0]
    
    return arparams, maparams, intercept
```

```python
class ARIMAPredictor:
    def __init__(self, p, d, q, arparams, maparams, intercept):
        self.p = p
        self.d = d
        self.q = q
        self.arparams = arparams
        self.maparams = maparams
        self.intercept = intercept
        
        self.history_raw = []
        self.history_diff = []
        self.residuals = []

    def update_history(self, value):
        if self.d > 0:
            if len(self.history_raw) > 0:
                diff_value = value - self.history_raw[-1]
                self.history_diff.append(diff_value)
            else:
                self.history_diff.append(value)
        else:
            self.history_diff.append(value)
        
        self.history_raw.append(value)
    
    def reverse_differencing(self, diff_pred):
        if self.d == 0:
            return diff_pred
        elif self.d == 1:
            if len(self.history_raw) == 0:
                return diff_pred
            else:
                return self.history_raw[-1] + diff_pred
        else:
            raise ValueError("This code only supports d = 0 or d = 1.")
    
    def predict(self, value):
        ar_contrib = np.dot(self.arparams, self.history_diff[-self.p:][::-1]) if len(self.history_diff) >= self.p else 0
        ma_contrib = np.dot(self.maparams, self.residuals[-self.q:][::-1]) if len(self.residuals) >= self.q else 0
        diff_pred = self.intercept + ar_contrib + ma_contrib
        prediction = self.reverse_differencing(diff_pred)
        self.update_history(value)
        residual = prediction - value
        self.residuals.append(residual)
        
        return prediction
```

```python
def read_file(filename):
    prices = []
    with open(filename, 'r') as file:
        content = file.read()
    prices = list(map(float, content.split()))
    return np.array(prices)

dataset1 = read_file('./FinalProject/data-preparation/2024-11-01.txt')
dataset2 = read_file('./FinalProject/data-preparation/2024-11-20.txt')

p, d, q = 2, 1, 1

arparams, maparams, intercept = get_arima_coefficients(dataset1, p, d, q)
predictor = ARIMAPredictor(p, d, q, arparams, maparams, intercept)
predictions = []

for value in dataset2:
    prediction = predictor.predict(value)
    predictions.append(prediction)

import matplotlib.pyplot as plt
x_values = list(range(len(predictions)))

print(arparams)
print(maparams)
print(intercept)
print(predictor.residuals)

plt.plot(x_values, dataset2, color='g')
plt.plot(x_values, predictions, color='r')
plt.ylim(210, 230)
plt.show()
```
