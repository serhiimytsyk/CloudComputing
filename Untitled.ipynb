```python
import yfinance as yf

# Get historical data for a stock
ticker = "AAPL"
data = yf.download(ticker, interval="1m", start="2024-11-14", end="2024-11-15")

# Calculate mid-price
data['Mid Price'] = (data['High'] + data['Low']) / 2

print(data[['High', 'Low', 'Mid Price']])
print(len(data))
print(type(data))
print(data.columns)
```

```python
mid_prices = data['Mid Price']
```

```python
import matplotlib.pyplot as plt
# Plot the Mid Price to visualize trends
plt.figure(figsize=(10, 6))
plt.plot(mid_prices, label="Mid Price")
plt.title("Mid Price Time Series")
plt.xlabel("Datetime")
plt.ylabel("Mid Price")
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform ADF test
result = adfuller(mid_prices)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")

if result[1] < 0.05:
    print("The data is stationary.")
else:
    print("The data is not stationary. Differencing is required.")
```

```python
# Differencing to make the data stationary
mid_prices_diff = mid_prices.diff().dropna()

# Recheck stationarity after differencing
result = adfuller(mid_prices_diff)
print(f"p-value after differencing: {result[1]}")
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Define the ARIMA model (example order: (p=2, d=1, q=2))
model = ARIMA(mid_prices[:-50], order=(2, 1, 2))
model_fit = model.fit()

# Print model summary
print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=50)
```

```python
import pandas as pd
# Get the last timestamp in the original data
last_date = mid_prices.index[-50]

# Generate future timestamps based on the original data's frequency
future_dates = pd.date_range(start=last_date, periods=len(forecast) + 1, freq='T')[1:]  # 'T' for minutes

# Assign the generated dates to the forecast index
forecast.index = future_dates

# Print the forecast with dates
print(forecast)

# Plot the forecast with actual data
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(mid_prices[:-50], label="Actual Prices")
plt.plot(forecast, label="Forecasted Prices", color='red')
plt.title("Mid Price Forecasting with Dates")
plt.xlabel("Datetime")
plt.ylabel("Mid Price")
plt.legend()
plt.show()
```
